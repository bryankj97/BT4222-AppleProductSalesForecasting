```python
import pandas as pd
import snscrape.modules.twitter
```

```python
#create json representation of sns tweet object
def unpack_tweet(s):
    return {'url': s.url,
            'date': s.date,
            'content': s.content,
            'id': s.id,
            'user': s.user,
            'outlinks': s.outlinks,
            'tcooutlinks': s.tcooutlinks,
            'replyCount': s.replyCount,
            'retweetCount': s.retweetCount,
            'likeCount': s.likeCount,
            'quoteCount': s.quoteCount,
            'conversationId': s.conversationId,
            'lang': s.lang,
            'source': s.source,
            'sourceUrl': s.sourceUrl,
            'sourceLabel': s.sourceLabel,
            'media': s.media,
            'retweetedTweet':s.retweetedTweet,
            'quotedTweet': s.quotedTweet,
            'mentionedUsers':s.mentionedUsers}


def unpack_list_of_tweets(list_s):
    return pd.DataFrame([unpack_tweet(x) for x in list_s])
```

```python
def get_tweets(start, end, topic, limit=10):
    tweets = []
    i = 0
    query = topic + ' lang:en since:' + str(start) + ' until:' + str(end) + " -filter:links"
    for tweet in snscrape.modules.twitter.TwitterSearchScraper(query).get_items():
        i+=1
        tweets.append(tweet)
        if i > limit:
            break
    return tweets
```

```python
def get_n_tweets_from_range(n, start, end, keyword):
    
    # sample equal number of tweets from each day from start to end date
    time = pd.DataFrame(pd.date_range(start, end,freq='5d',).shift(1) , pd.date_range(start, end,freq='5d',))
    time = time.reset_index()
    time.columns = ['start','end']
    time.start = time.start.astype('str')
    time.end = time.end.astype('str')
    
    numperday = n/time.shape[0]
    
    print("Collecting " + str(numperday) + " tweets per period - Total " + str(time.shape[0]) + " Periods")
    df = pd.DataFrame()

    for i, r in time.iterrows():
        tweets = get_tweets(r.start, r.end, keyword, numperday)
        df = pd.concat([df, unpack_list_of_tweets(tweets)])
        print(r.start + " Scraped")

    
    return df.reset_index()
```

```python
# start with 20k tweets 

start = '2011-12-31'
end = '2018-12-31'

df = get_n_tweets_from_range(20000, start, end, "iphone")
```
